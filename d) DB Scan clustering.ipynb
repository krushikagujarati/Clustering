Read dataset

```python
import pandas as pd

# Load the dataset
file_path = 'world_development_data_imputed.csv'
data = pd.read_csv(file_path)

# Displaying the first few rows of the dataset to understand its structure
data.head()

```

Data Preparation

```python
# Selecting relevant columns (example columns, adjust according to need)
selected_columns = ['PopTotal', 'GDP', 'LifeExpBirth', 'UrbanPopGrowth%', 'GNIAtlas']
cluster_data = data[selected_columns]
```


Standardizing the data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cluster_data_scaled = scaler.fit_transform(cluster_data)
```

Setting Up PyCaret Environment

```python
pip install pycaret
```

```python
from pycaret.clustering import *
cluster_setup = setup(data=cluster_data, normalize=True, session_id=123)
```

```python
# Importing necessary libraries
from pycaret.clustering import *

# Performing DBSCAN Clustering
dbscan_model = create_model('dbscan', eps=0.5, min_samples=5)  # Adjust eps and min_samples as needed

```

Analyzing the Clusters

```python
dbscan_results = assign_model(dbscan_model)

# Displaying the first few rows of the results
dbscan_results.head()
```

Plotting Model

```python
from pycaret.clustering import plot_model

plot_model(dbscan_model, plot='cluster')
```
